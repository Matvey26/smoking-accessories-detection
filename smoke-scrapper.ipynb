Но зато фотки для датасета отличные

```python
BASE_URL = 'https://baraholka.maxvg.com.ua/iqos-glo-cistemi-nagrjvu-tabaku'
# на этом сайте есть кнопки для переключения страниц, вот краткое описание для них:
# тег: <a id="cat-link" rel="число"> <span> Название кнопки </span> </a>
# мне нужны rel in [96, 107, 105, 97, 98, 101, 102] (это категории)

# на каждой странице (при нажатии на кнопку) есть куча фоток и снизу тоже фотки для переключения подстраниц этой страницы
# фотки можно найти по такому описанию:
# <a class="img-link"> <img src="НУЖНАЯ НАМ ССЫЛКА ДЛЯ СКАЧИВАНИЯ ФОТКИ"> </a>

# найти кнопки для переключения подстраниц можно по ссылке:
# BASE_URL/<number>
```

```python
import os
import time
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By

# Настройки
CATEGORIES = [96, 107, 105, 97, 98, 101, 102]
DOWNLOAD_FOLDER = 'downloaded_images'

# Создаем папку для скачивания
if not os.path.exists(DOWNLOAD_FOLDER):
    os.makedirs(DOWNLOAD_FOLDER)

# Инициализация драйвера
driver = webdriver.Chrome()
driver.get(BASE_URL)

time.sleep(3)

def download_images_from_page(url):
    """Скачивает все фотографии с указанной страницы."""
    driver.get(url)
    time.sleep(2)  # Ждем загрузки страницы

    img_links = driver.find_elements(By.CSS_SELECTOR, 'a.img-link img')
    downloaded = 0
    for img in img_links:
        img_url = img.get_attribute('src')
        if img_url:
            try:
                img_data = requests.get(img_url).content
                img_name = os.path.join(DOWNLOAD_FOLDER, f"{time.time()}.jpg")
                with open(img_name, 'wb') as f:
                    f.write(img_data)
                print(f"Скачано: {img_name}")
                downloaded += 1
            except Exception as e:
                print(f"Ошибка при скачивании {img_url}: {e}")
    return downloaded

def process_category(category_rel):
    """Обрабатывает одну категорию: переключает подстраницы и скачивает фотографии."""
    # Находим ссылку на категорию
    category_link = driver.find_element(By.CSS_SELECTOR, f'a[rel="{category_rel}"]')
    base_category_url = category_link.get_attribute('href')
    print(f"Обработка категории {category_rel}: {base_category_url}")

    page = 1
    total_downloaded = 0
    max_images = 60  # Максимальное количество фотографий для скачивания

    while total_downloaded < max_images:
        page_url = f"{base_category_url}/{page}" if page > 1 else base_category_url
        print(f"Обработка страницы {page}: {page_url}")

        downloaded = download_images_from_page(page_url)
        if downloaded == 0:  # Если не скачали ни одной фотографии, значит страниц больше нет
            break

        total_downloaded += downloaded
        page += 1

# Основной цикл
for category in CATEGORIES:
    process_category(category)

# Закрываем драйвер
driver.quit()
print("Скрипт завершен.")
```
